### Create feather per 1000 User_id 

```python
# Drop duplicate for count users
user_id = data['user_id'].drop_duplicates()
user_id.reset_index(inplace=True,drop=True)
```

```python
print(user_id)
print(len(user_id))
```

```python
# make user_dic (per 1000)
user_dic = {}
count = 0
for i in range(len(user_id)):
    if i%1000 == 0:
        count +=1
    if count not in user_dic:
        user_dic[count] = [user_id[i]]
    else:
        user_dic[count].append(user_id[i])
```

```python
# Num of Keys
print(len(user_dic.keys()))
print(len(user_dic[1]))
print(len(user_dic[394]))
```

```python
# Create feather file
for i in range(1,len(user_dic.keys())+1):
    data[data['user_id'].isin(user_dic[i])].reset_index(drop=True).to_feather(f'C:\\Users\\lmj35\\Desktop\\벌쿠형 플젝\\user_id_per_1000\\user_dic_{i}.feather')
```
