## Load file list

```python
contentFiles= []
for (dirpath, dirnames, filenames) in os.walk('./content_id_per_50'):
    contentFiles.extend(filenames)
```

```python
userFiles= []
for (dirpath, dirnames, filenames) in os.walk('./user_id_per_1000'):
    userFiles.extend(filenames)
```

## Build dictionary for fileName:itemID

```python
userFiles = userFiles[371:]
```

```python
userFiles
```

```python
userFileIdDic = {}
```

```python
for uf in userFiles:
    df = pd.read_feather('user_id_per_1000/'+uf)
    df = df.sort_values(by=['user_id','timestamp'], axis=0)
    priorQuestionChange(df)
    lectureLabelChange(df)
    uids = list(set(df.user_id))
    userFileIdDic[uf] = uids
    
    #extract feature
    addAccumulateAcc(df, 0.9)
    addWindowAcc(df, 20)
    df.reset_index(inplace=True)
    df.drop(['index'], axis=1, inplace=True)
    df.to_feather('user_id_feature/'+uf)
    
    print(uf)
    
```

```python
contentsFileIdDic = {}
```

```python
for cf in contentFiles:
    df = pd.read_feather('content_id_per_50/'+cf)
    df = df.sort_values(by=['content_id','timestamp'], axis=0)
    priorQuestionChange(df)
    lectureLabelChange(df)
    qids = list(set(df.content_id))
    contentsFileIdDic[cf] = qids
    
    #extract feature
    addAccumulateCountDiff(df)
    df.reset_index(inplace=True)
    df.drop(['index'], axis=1, inplace=True)
    df.to_feather('content_id_feature/'+cf)
    
    print(cf)
```
